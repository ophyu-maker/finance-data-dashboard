```python
employee_clean = employeeEarn_df.copy()

employee_clean.columns = (
    employee_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
employee_clean.columns.to_list()
```

```python
money_cols = [
    "regular",
    "retro",
    "other",
    "overtime",
    "injured",
    "detail",
    "quinn_education",
    "total_gross"
]
```

```python
def clean_money_column(series):
    return (
        series
        .astype(str)
        .str.strip()
        .str.replace(",", "", regex=False)
        .str.replace("$", "", regex=False)
        .replace(["nan", "None", "", " "], np.nan)
        .pipe(pd.to_numeric, errors="coerce")
    )

for col in money_cols:
    employee_clean[col] = clean_money_column(employee_clean[col])

# Fill missing pay component columns with 0
pay_component_cols = [
    "regular",
    "retro",
    "other",
    "overtime",
    "injured",
    "detail",
    "quinn_education"
]
```

```python
employee_clean[pay_component_cols] = employee_clean[pay_component_cols].fillna(0)
```

```python
text_cols = ["name", "department_name", "title"]

for col in text_cols:
    employee_clean[col] = employee_clean[col].astype(str).str.strip()
```

```python
employee_clean["postal"] = employee_clean["postal"].astype(str).str.strip()
```
