```python
df_unique_departments = df.drop_duplicates(subset=['DEPARTMENT_NAME'], keep='first')
df_unique_departments = df_unique_departments['DEPARTMENT_NAME']
df_unique_departments
```

```python
months = pd.date_range(start="2025-01-01", end="2025-12-01", freq="MS")
```

```python
def assign_budget_category(dept):
    dept_lower = dept.lower()

    if "police" in dept_lower:
        return "very_large"
    elif "fire" in dept_lower:
        return "very_large"
    elif "school" in dept_lower or "elementary" in dept_lower or "academy" in dept_lower or "high" in dept_lower:
        return "school"
    elif "public works" in dept_lower or "transportation" in dept_lower or "parks" in dept_lower:
        return "large"
    elif "technology" in dept_lower or "finance" in dept_lower or "inspectional" in dept_lower:
        return "medium"
    elif "office" in dept_lower or "commission" in dept_lower or "board" in dept_lower or "strategy" in dept_lower:
        return "small"
    else:
        return "medium_small"

def generate_base_budget(category):
    if category == "very_large":
        return np.random.uniform(8_000_000, 18_000_000)
    elif category == "school":
        return np.random.uniform(500_000, 2_500_000)
    elif category == "large":
        return np.random.uniform(3_000_000, 8_000_000)
    elif category == "medium":
        return np.random.uniform(1_000_000, 3_000_000)
    elif category == "medium_small":
        return np.random.uniform(300_000, 1_200_000)
    elif category == "small":
        return np.random.uniform(80_000, 500_000)
```

```python
budget_rows = []

for dept in df_unique_departments:
    category = assign_budget_category(dept)
    base_annual_budget = generate_base_budget(category)
    base_monthly_budget = base_annual_budget / 12

    for month in months:
        # Add monthly variation between -5% and +5%
        monthly_variation = np.random.uniform(0.95, 1.05)
        budget_amount = round(base_monthly_budget * monthly_variation, 2)

        budget_rows.append({
            "month": month.strftime("%Y-%m"),
            "department_name": dept,
            "budget_category": category,
            "budget_amount": budget_amount
        })

budgets_df = pd.DataFrame(budget_rows)

budgets_df.head()
```

```python
budgets_df.to_csv("budgets.csv", index=False)
```

Budget data was synthetically generated for portfolio demonstration purposes using rule-based department categories and randomized monthly variation.
